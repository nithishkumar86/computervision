```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("thala.jpg")
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hsv_image=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray_image)
plt.title("gray_scale iame")

plt.subplot(1,2,2)
plt.imshow(hsv_image)
plt.title("hsv_image")
```

```python
img=cv2.imread("thala.jpg")
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
ret,thresh=cv2.threshold(gray_image,128,255,cv2.THRESH_BINARY)
```

```python
thresh
```

```python
counter,heirarch=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
```

```python
len(counter)
```

```python
counter
```

```python
counter_image=cv2.drawContours(img,counter,-1,(0,0,255),-1)
```

```python
plt.imshow(counter_image)
```
