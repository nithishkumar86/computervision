```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from PIL import Image
import cv2
```

```python
datagen=ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=0.2,
    fill_mode="reflect"
)
```

```python
img=image.load_img("D:\\images\\dhoni\\dhoni.jpg",target_size=(224,224))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
print(type(img))
```

```python
img=image.img_to_array(img)
```

```python
img=img.reshape(1,224,224,3)
```

```python
i=0
for batch in datagen.flow(img,
                          batch_size=1,
                          save_to_dir="D:\\images\\dhoni",
                          save_prefix="dhoni",
                          save_format="jpg"):
    i+=1
    if i>10:
        break
```
