```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('thala.jpg')
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_img)
```

```python
kernelsize=[(3,3),(9,9),(15,15)]
```

```python
for i in kernelsize:
    print(i)
```

```python
kernelsize[1]
```

#a gussian blur is a technique to reduce noise 

#a larger kernel size will result in more blurrring

#while a smaller kernel size will cause less blurring

#Decreaing it (eg:(3,3)will reduce  the blurring effect)


Syntax: cv2. GaussianBlur(image, shapeOfTheKernel, sigmaX )


Image– the image you need to blur

shapeOfTheKernel– The shape of the matrix-like 3 by 3 / 5 by 5

sigmaX– The Gaussian kernel standard deviation which is the default set to 0



```python

blurred_img1=cv2.GaussianBlur(rgb_img,kernelsize[0],0)
blurred_img2=cv2.GaussianBlur(rgb_img,kernelsize[1],0)
blurred_img3=cv2.GaussianBlur(rgb_img,kernelsize[2],0)
blurred_img4=cv2.GaussianBlur(blurred_img3,kernelsize[2],0)
```

```python
kernel_2=np.array([[-1,-1,-1],[2,9,-1],[-1,-1,-1]])
```

#the sharpening effect is controlled by the values in the covolution kernel applied with

#cv2.filter2D() applied with cv2.fiter2d the larger the center value the sharper the result

#Syntax: filter2D(sourceImage, ddepth, kernel)

```python
sharpened=cv2.filter2D(rgb_img,0,kernel_2)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,5,1)
plt.imshow(blurred_img1)
plt.title("blurred_img1")

plt.subplot(1,5,2)
plt.imshow(blurred_img2)
plt.title("blurred_img2")

plt.subplot(1,5,3)
plt.imshow(blurred_img3)
plt.title("blurred_img3")

plt.subplot(1,5,4)
plt.imshow(blurred_img4)
plt.title("blurred_img4")

plt.subplot(1,5,5)
plt.imshow(sharpened)
plt.title("sharpened")

```
