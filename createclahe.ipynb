#clahe->contrast limited adaptive histogram equalization for improving the contrast

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('lung.jpg')
```

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

#createCLAHE - can be done only through gray scale image

#clipLimit - threshold for contrast limiting

#tilegridsize - size of grid for histogram equalization input image will be divided into
                into equally size recatangular images  

```python
clahe=cv2.createCLAHE(clipLimit=5,tileGridSize=(8,8))
```

```python
clahe_img=clahe.apply(gray_img)
```

```python
plt.Figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.title('rgb_img')

plt.subplot(1,2,2)
plt.imshow(clahe_img)
plt.title('clahe_img')
```

#clahe using the same image LAB

```python
img1=cv2.imread('lung.jpg')
```

```python
lab=cv2.cvtColor(img1,cv2.COLOR_BGR2LAB)
```

```python
l,a,b=cv2.split(lab)
```

```python
plt.imshow(l)
```

```python
#When applying CLAHE, there are two parameters to be remembered:
#clipLimit – This parameter sets the threshold for contrast limiting. The default value is 40. 
#tileGridSize – This sets the number of tiles in the row and column. By default this is 8×8. It is used while the image is divided into tiles for applying CLAHE.  
clahe=cv2.createCLAHE(clipLimit=0.3,tileGridSize=(8,8))
```

```python
cla=clahe.apply(l)
```

```python
lab_image=cv2.merge([cla,a,b])
```

```python
clahe_rgb_img=cv2.cvtColor(lab_image,cv2.COLOR_LAB2BGR)
```

```python
plt.Figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(lab)
plt.title('lab_img')

plt.subplot(1,2,2)
plt.imshow(clahe_rgb_img)
plt.title('clahe_rgb_img')
```
