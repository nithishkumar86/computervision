#opencv using cvtCOLOR

#cv2.cvtColor(img_name,cv2.COLOR_BGR2GRAY,cv2.COLOR_BGR2HSV,cv2.COLOR_BGR2RGB)

```python
img=cv2.imread("D:\\image\\thala.jpg")
```

```python
cvt_img1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cvt_img2=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cvt_img3=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
cv2.imwrite("D:\\image\\thala2.jpg",cvt_img1)
cv2.imwrite("D:\\image\\thala3.jpg",cvt_img2)
cv2.imwrite("D:\\image\\thala4.jpg",cvt_img3)
```

#cv2.bitwise_not()

#bitwise_not will flips the value of image

#all pixel value that are greater than will set 0

#all pixel value that are equal to zero is set 255

```python
img=cv2.imread("D:\\image\\thala.jpg")
```

```python
res=cv2.bitwise_not(img)
```

```python
cv2.imwrite("D:\\image\\thala5.jpg",res)
```

#add() this method is used add two images 

```python
img1=cv2.imread("D:\\image\\thala.jpg")
img2=cv2.imread("D:\\image\\thala3.jpg")
```

```python
res1=cv2.add(img1,img2)
```

```python
cv2.imwrite("D:\\image\\thala6.jpg",res1)
```

#addWeighted(img,(transparency_level(0-1)),img2,(transparency_level(0-1)),thickness

```python
img1=cv2.imread("D:\\image\\thala.jpg")
img2=cv2.imread("D:\\image\\thala3.jpg")
```

```python
res2=cv2.addWeighted(img1,0.5,img2,0.5,100)
```

```python
cv2.imwrite("D:\\image\\thala7.jpg",res2)
```

#rectangle(img,start_point(x,y),stop_point(x,y),color(bgr),thickness)

```python
img=cv2.imread("D:\\image\\thala.jpg")
```

```python
res2=cv2.rectangle(img,(10,10),(10,10),(255,255,255),100)
```

```python
cv2.imwrite("D:\\image\\thala8.jpg",res2)
```
