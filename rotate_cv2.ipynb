cv2.rotate()- rotate the image into only three angles these are inbuilt

cv2.ROTATE_90

CV2.ROTATE_90_CLOCKWISE

cv2.ROTATE_90_COUNTERCLOCKWISE


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('thala.jpg')
rgb_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
rotate_180=cv2.rotate(rgb_image,cv2.ROTATE_180)
rotate_90c=cv2.rotate(rgb_image,cv2.ROTATE_90_CLOCKWISE)
rotate_90cc=cv2.rotate(rgb_image,cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,4,1)
plt.imshow(rgb_image)
plt.title('original_image')

plt.subplot(1,4,2)
plt.imshow(rotate_180)
plt.title('ROTATE_180')

plt.subplot(1,4,3)
plt.imshow(rotate_90c)
plt.title('ROTATE_90_CLOCKWISE')

plt.subplot(1,4,4)
plt.imshow(rotate_90cc)
plt.title('ROTATE_90_COUNTERCLOCKWISE')
```

custom rotation

getRotationMatrix2D() to rotate an image in user required angle

wrapAffine(original_image,centered_image,(size)) apply the roatation on the image

```python
#step1-load the image and convert the image into rgb
img2=cv2.imread('thala.jpg')
rgb_image2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
#step2-find the center coordinate of the image(height/2,width/2) 
h,w=rgb_image2.shape[0:2]
```

```python
center_image=(w/2,h/2)
```

```python
#step-to rotate the image in the  given the angle by getRotationMatrix2D
rotate_matrix=cv2.getRotationMatrix2D(center_image,angle=300,scale=1)
```

```python
rotate_image=cv2.warpAffine(rgb_image2,rotate_matrix,(146,146))
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(rgb_image2)
plt.title('original_image')

plt.subplot(1,2,2)
plt.imshow(rotate_image)
plt.title('rotate_image')
```

#flip in images

flipcode=0 =>flips the image vertically (around the x-axis)

flipcode=1 =>flips the image horizontally (around the y-axis)

flipcode=-1 =>flips the image vertically and horizontally (180 degree)

```python
img3=cv2.imread('thala.jpg')
rgb_img4=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
```

```python
flip_0=cv2.flip(rgb_img4,flipCode=0)
flip_1=cv2.flip(rgb_img4,flipCode=1)
flip_neg_1=cv2.flip(rgb_img4,flipCode=-1)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,4,1)
plt.imshow(rgb_img4)
plt.title('original_image')

plt.subplot(1,4,2)
plt.imshow(flip_0)
plt.title('flip_0')

plt.subplot(1,4,3)
plt.imshow(flip_1)
plt.title('flip_1')

plt.subplot(1,4,4)
plt.imshow(flip_neg_1)
plt.title('flip_neg_1')
```

#adding noise to image

```python
image2=cv2.imread('thala.jpg')
rgb_image3=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
```

```python
import random
import numpy as np
```

```python
print(rgb_image3.shape)
```

```python
noisy_image=np.random.normal(80,140,rgb_image3.shape)
```

```python
print(noisy_image)
```

```python
len(noisy_image)
```

```python
noisy_image=rgb_image3+noisy_image
```

```python
noisy_image1=np.clip(noisy_image,a_min=0,a_max=255).astype(np.uint8)
```

```python
noiseimage=cv2.cvtColor(noisy_image1,cv2.COLOR_BGR2RGB)
```

```python
plt.Figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.imshow(rgb_image3)
plt.title("original_image")

plt.subplot(1,2,2)
plt.imshow(noiseimage)
plt.title("rgb_image3")
```
